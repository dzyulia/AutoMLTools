```python
import pandas as pd
from autogluon.timeseries import TimeSeriesDataFrame, TimeSeriesPredictor
```

```python
#load the data and print it
df = pd.read_csv("C:/Users/dzyyu/AutoMLTools/AutoML-Tools/AutoGluon/Datasets/time_series2.csv")
df.head()
```

```python
# convert from DataFrame into TimeSeriesDataFrame format.
train_data = TimeSeriesDataFrame.from_data_frame(
    df,
    id_column="item_id",
    timestamp_column="timestamp"
)
train_data.head()
```

```python
# predicts future values of multiple related time series
predictor = TimeSeriesPredictor(
    prediction_length=4,
    path="autogluon-m4-hourly",
    target="target",
    eval_metric="MASE"
)

predictor.fit(
    train_data,
    presets="medium_quality",
    time_limit=300
)
```

```python
# prediction of next time-series
predictions = predictor.predict(train_data)
predictions.head()
```

```python
#print leaderboard of models
leaderboard = predictor.leaderboard()
```
