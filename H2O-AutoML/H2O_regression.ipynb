```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
#load dataset
data = h2o.import_file('C:/Users/dzyyu/AutoMLTools/AutoML-Tools/H2O-AutoML/Datasets/garments_worker_productivity_regression.csv')
```

```python
#dataset description
data.describe()
```

```python
output = 'actual_productivity'
train, test = data.split_frame(ratios=[0.8])
features = ['date', 'quarter', 'department', 'day', 'team', 'targeted_productivity', 'smv', 'wip', 'over_time',
                'incentive', 'idle_time', 'idle_men', 'no_of_style_change', 'no_of_workers']
aml = H2OAutoML(max_runtime_secs=600, seed = 1)
aml.train(x = features, y = output, training_frame = train)
```

```python
#get leaderboard of best models
lb = aml.leaderboard
print(lb)
lb.head(rows=lb.nrow)
#save best model
model_path = h2o.save_model(model=aml.leader, path="C:/Users/dzyyu/Downloads/dataset", force=True)
performance = aml.leader.model_performance(test_data=test)
print(performance.mse())
print(model_path)
```

```python
#load model
saved_model = h2o.load_model(model_path)

# Use the model to make predictions
predictions = saved_model.predict(test)
```

```python
print (predictions)
```
